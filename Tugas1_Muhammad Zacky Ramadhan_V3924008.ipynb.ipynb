```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter distribusi Poisson
lambda_rate = 8  # rata-rata pelanggan per jam
total_hours = 12 * 30  # 12 jam per hari selama 30 hari

# Simulasi kedatangan pelanggan
simulated_data = np.random.poisson(lam=lambda_rate, size=total_hours)

# Histogram dari data simulasi
plt.figure(figsize=(10, 6))
plt.hist(simulated_data, bins=range(min(simulated_data), max(simulated_data)+2), 
         align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title("Distribusi Frekuensi Jumlah Kedatangan Pelanggan per Jam")
plt.xlabel("Jumlah Kedatangan Pelanggan")
plt.ylabel("Frekuensi")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Hitung probabilitas teoritis
p_zero_customers = poisson.pmf(0, mu=lambda_rate)
p_more_than_10 = 1 - poisson.cdf(10, mu=lambda_rate)

# Hitung probabilitas empiris dari simulasi
empirical_p_zero = np.mean(simulated_data == 0)
empirical_p_more_than_10 = np.mean(simulated_data > 10)

# Distribusi teoritis untuk perbandingan
x_values = np.arange(0, max(simulated_data)+1)
theoretical_probs = poisson.pmf(x_values, mu=lambda_rate)

# Frekuensi dari simulasi (empiris)
simulated_freqs = np.array([np.mean(simulated_data == x) for x in x_values])

# Plot perbandingan teoritis vs empiris
plt.figure(figsize=(10, 6))
plt.plot(x_values, theoretical_probs, 'o-', label='Teoritis (Poisson)', color='red')
plt.bar(x_values, simulated_freqs, alpha=0.6, label='Simulasi (Empiris)', color='blue')
plt.title("Perbandingan Distribusi Teoritis vs Simulasi")
plt.xlabel("Jumlah Kedatangan Pelanggan")
plt.ylabel("Probabilitas")
plt.legend()
plt.grid(True)
plt.show()

# Tampilkan hasil probabilitas
print("Probabilitas Teoritis P(X=0):", p_zero_customers)
print("Probabilitas Teoritis P(X>10):", p_more_than_10)
print("Probabilitas Empiris P(X=0):", empirical_p_zero)
print("Probabilitas Empiris P(X>10):", empirical_p_more_than_10)
```
